For `score_rating` the average score is of course over 1 because [they're good dogs brent](https://knowyourmeme.com/memes/theyre-good-dogs-brent). Let's look at the dogs with the highest score rating (besides the American dog). We will look at which breeds of dogs have the highest scores

```python
breed_ratings = df_twitter.groupby('p1')['score_rating'].mean()
```

```python
breed_ratings = breed_ratings.sort_values(ascending = False)
```

```python
plt.style.use('seaborn')
breed_ax = breed_ratings.plot(kind='bar', figsize=(25,8))

breed_ax.set_title("Highest Rated Dog Breeds on WeRateDogs", fontsize = 24)
breed_ax.set_xlabel("")
```

With the graph above, we can see the most popular breeds in order from best to worst. The Clumber seems to have a much bigger lead then the rest of the group. Lets look at this breed
![clumber](https://s3.amazonaws.com/cdn-origin-etr.akc.org/wp-content/uploads/2017/11/07135844/Clumber-Spaniel-On-White-011.jpg)

Just for fun, lets look at the most popular locations of the users that get tagged in WeRateDogs.

```python
plt.style.use('seaborn')
location_ax = user_extract.location.value_counts()[1:99].plot(kind = "bar", figsize = (25,8))

location_ax.set_title("Most Popular Locations from WeRateDogs Tagged Users", fontsize = 24)
location_ax.set_xlabel("")
```

Since many locations were not given from the users, and anything can be put in the location field on your profile, the data was a bit inconsistent. That said, it looks like out of the locations that were collected, Los Angeles, CA was the most popular location. Since there were so many different strings in the `locations` column, they would all be hard to display, so these are the top 100.
